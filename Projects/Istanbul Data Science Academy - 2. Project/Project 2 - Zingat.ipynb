```python
from urllib.request import urlopen
import requests
from bs4 import BeautifulSoup as bts
import pandas as pd
import re
import time
import sys
from datetime import datetime
from datetime import date
```

```python
def getAndParseURL(url):
    #result = requests.get(url,headers = {"User-Agent": 'Mozilla/5.0'})
    #result = requests.get(url,headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.103 Safari/537.36'})
    result = requests.get(url,headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.103 Safari/537.36'})
    soup = bts(result.text, 'html.parser')
    return soup

```

```python
page_links = list()
#Find the total number of the houses in Kadıköy
town_links = ['https://www.zingat.com/kadikoy-caferaga-satilik','https://www.zingat.com/kadikoy-osmanaga-satilik',
             'https://www.zingat.com/kadikoy-rasimpasa-satilik','https://www.zingat.com/kadikoy-kosuyolu-satilik',
             'https://www.zingat.com/kadikoy-dumlupinar-satilik','https://www.zingat.com/kadikoy-fikirtepe-satilik',
             'https://www.zingat.com/kadikoy-egitim-satilik','https://www.zingat.com/kadikoy-feneryolu-satilik',
             'https://www.zingat.com/kadikoy-zuhtupasa-satilik','https://www.zingat.com/kadikoy-fenerbahce-satilik',
             'https://www.zingat.com/kadikoy-caddebostan-satilik','https://www.zingat.com/kadikoy-goztepe-satilik',
             'https://www.zingat.com/kadikoy-merdivenkoy-satilik','https://www.zingat.com/kadikoy-sahrayi-cedit-satilik',
             'https://www.zingat.com/kadikoy-19-mayis-satilik','https://www.zingat.com/kadikoy-erenkoy-satilik',
             'https://www.zingat.com/kadikoy-suadiye-satilik','https://www.zingat.com/kadikoy-kozyatagi-satilik',
             'https://www.zingat.com/kadikoy-bostanci-satilik','https://www.zingat.com/kadikoy-acibadem-satilik',
              'https://www.zingat.com/kadikoy-hasanpasa-satilik']
for i in town_links:
    print(i)
    number = str()
    house_number = str()
    html = getAndParseURL(i)
    for house in html.find_all("div",{"class":"page-title"}):
        number = house.span.text

    num = [int(x) for x in house_number.split() if x.isdigit()] 
    for j in number:
        if j.isdigit():
            house_number += str(j)
    house_number = int(house_number)

    #Find the total page number
    page_number = int()
    if house_number%21 == 0:
        page_number = house_number//21
    else:
        page_number = house_number//21+1
    for j in range(1,page_number+1):
        page_links.append(i + "?page=" + str(j))
    time.sleep(30)
    
```

```python
house_links = list()
for pgn in page_links:
    print(pgn)
    html = getAndParseURL(pgn)
    for link in html.find_all('a', class_='zl-card-inner'):
        house_links.append(link['href'])
```

```python
listing_names = list()
listing_prices = list()
location = list()
netm2 = list()
bedroom_number = list()
building_floor_number = list()
floor = list()
heating_system = list()
age = list()
situation = list()
seller = list()
title_dead_status = list()
credit = list()
rent_income = list()
```

```python
nonscraped_house_links = list()
```

```python
import random


for house in range(len(house_links)):
    print(house, house_links[house])
    try:
        html = getAndParseURL('https://www.zingat.com'+house_links[house])
        try:
            temp = list()
            for link in html.find_all("h1",{"itemprop":"name"}):
                temp.append(link.text)
            listing_names.append(temp[0])
        except:
            continue

        temp = list()
        price = str()
        for link in html.find_all("div",{"class":"listing-price"}):
            temp.append(link.text)

        for j in temp[0]:
            if j.isdigit():
                price += str(j)
        listing_prices.append(int(price))

        temp = list()
        for link in html.find_all("a",{"class":"location-link"}):
            temp.append(link.text)
        location.append(temp[2].split()[0].replace(",",""))

        temp_keys = list()
        temp_values = list()

        for link in html.find_all("span",{"class":"col-md-6"}):
            temp_values.append(link.text)
        for link in html.find_all("strong",{"class":"col-md-6"}):
            temp_keys.append(link.text)
        temp_dict = dict(zip(temp_keys,temp_values))

        try:
            net_m2 = str()
            for j in temp_dict['Net m²']:
                if j.isdigit():
                    net_m2 += str(j)
            try:
                netm2.append(int(net_m2))  
            except:
                netm2.append(int(net_m2[0:-1])) 
        except:
            netm2.append(None)

        try:
            bedroom = str()
            for j in temp_dict['Bölüm/Oda Sayısı']:
                if j.isdigit():
                    bedroom += str(j)
            bedroom_number.append(int(bedroom))
        except:
            try:
                bedroom = str()
                for j in temp_dict['Oda Sayısı']:
                    if j.isdigit():
                        bedroom += str(j)
                bedroom_number.append(int(bedroom))
            except:
                bedroom_number.append(None)

        try:
            building_floor = str()
            for j in temp_dict['Binadaki Kat Sayısı']:
                if j.isdigit():
                    building_floor += str(j)
            building_floor_number.append(int(building_floor))
        except:
            building_floor_number.append(None)

        try:
            floor_str = str()
            for i in temp_dict['Bulunduğu Kat'].split()[:-1]:
                floor_str += str(i)
                floor_str += " "
            floor_str += temp_dict['Bulunduğu Kat'].split()[-1]
            floor.append(floor_str)
        except:
            floor.append(None)

        try:
            heating = str()
            for i in temp_dict['Isıtma Tipi'].split()[:-1]:
                heating += str(i)
                heating += " "
            heating += temp_dict['Isıtma Tipi'].split()[-1]
            heating_system.append(heating)
        except:
            heating_system.append(None)

        try:
            age.append(temp_dict['Bina Yaşı'].split()[0])
        except:
            age.append(None)

        try:
            situation.append(temp_dict['Kullanım Durumu'].split()[0])
        except:
            situation.append(None)

        try:
            seller.append(temp_dict['Kimden'].split()[0])
        except:
            seller.append(None)

        try:
            title_dead = str()
            for i in temp_dict['Tapu Durumu'].split()[:-1]:
                title_dead += str(i)
                title_dead += " "
            title_dead += temp_dict['Tapu Durumu'].split()[-1]
            title_dead_status.append(title_dead)
        except:
            title_dead_status.append(None)

        try:
            credit.append(temp_dict['Krediye Uygun'].split()[0])
        except:
            credit.append(None)

        try:
            rent_str = str()
            for j in temp_dict['Kira Getirisi']:
                if j.isdigit():
                    rent_str += str(j)
            rent_income.append(int(rent_str))
        except:
            rent_income.append(None)

        time.sleep(random.randint(1,30))
    except requests.exceptions.ConnectionError:
        print("ConnectionError")
        nonscraped_house_links.append(house_links[house])
        
```

```python
import pandas as pd


df = pd.DataFrame(list(zip(listing_names, listing_prices, location, netm2, bedroom_number, building_floor_number,
                           floor, heating_system, age, situation, seller, title_dead_status, credit,rent_income)),
                  columns=["listing_names", "listing_prices", "location", "netm2", "bedroom_number",
                           "building_floor_number", "floor", "heating_system", "age", "situation", "seller", 
                           "title_dead_status", "credit","rent_income"])
df
```
