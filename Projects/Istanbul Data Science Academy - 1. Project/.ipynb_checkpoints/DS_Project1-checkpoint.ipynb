```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime,timedelta
```

```python
#13 weeks data scraping with the beginning date 05/03/2022
start_date = pd.Timestamp("Saturday, March 05, 2022 00:00:00")
date = start_date
url = "http://web.mta.info/developers/data/nyct/turnstile/turnstile_{}.txt"

source = []

for i in range(13):
    file_name = str(date.year)[2:4] + str(date.month).zfill(2) + str(date.day).zfill(2)
    date += pd.Timedelta(days=7)
    file_name = url.format(file_name)
    source.append(pd.read_csv(file_name))
```

```python
#Save the scraped data as dataframe
df = pd.concat(source)
df.info() # View dataframe content
df.shape # View dataframe's line and column details
```

```python
df.head()
```

```python
df.columns
```

```python
#Fix a column name
df.rename(columns={'EXITS                                                               ': 'EXITS'},inplace=True)
```

```python
df.columns
```

```python
#Remove blanks in data
df['C/A'] = df['C/A'].str.strip()
df['UNIT'] = df['UNIT'].str.strip()
df['SCP'] = df['SCP'].str.strip()
df['STATION'] = df['STATION'].str.strip()
df['LINENAME'] = df['LINENAME'].str.strip()
df['DIVISION'] = df['DIVISION'].str.strip()
df['DATE'] = df['DATE'].str.strip()
df['TIME'] = df['TIME'].str.strip()
df['DESC'] = df['DESC'].str.strip()
```

```python
#Remove duplicates
df = df.drop_duplicates(keep = 'first')
```

```python
#Drop unnecessary columns
df = df.drop(["DIVISION","LINENAME"],axis=1)
```

```python
df.head()
```

```python
#Drop NaN's row
df = df.dropna()
```

```python
#Create a datetime columns
df['DATETIME'] = df['DATE'] + " " + df['TIME']
df['DATETIME'] = pd.to_datetime(df['DATETIME'],format='%m/%d/%Y %H:%M:%S')
```

```python
df.head()
```

```python
#Create a day column
df['DAY'] = df['DATETIME'].dt.day_name()
```

```python
df.head()
```

```python
#Create a yeardate column
df['YEARDATE'] = pd.DatetimeIndex(df['DATE']).year
```

```python
df.head()
```

```python
#Create time range column
def conditions(s):
    if (s[0:2] == '00') or (s[0:2] == '01') or (s[0:2] == '02') or (s[0:2] == '03'):
        return '00:00 - 04:00'
    elif (s[0:2] == '04') or (s[0:2] == '05') or (s[0:2] == '06') or (s[0:2] == '07'):
        return '04:00 - 08:00'
    elif (s[0:2] == '08') or (s[0:2] == '09') or (s[0:2] == '10') or (s[0:2] == '11'):
        return '08:00 - 12:00'
    elif (s[0:2] == '12') or (s[0:2] == '13') or (s[0:2] == '14') or (s[0:2] == '15'):
        return '12:00 - 16:00'
    elif (s[0:2] == '16') or (s[0:2] == '17') or (s[0:2] == '18') or (s[0:2] == '19'):
        return '16:00 - 20:00'
    elif (s[0:2] == '20') or (s[0:2] == '21') or (s[0:2] == '22') or (s[0:2] == '23'):
        return '20:00 - 00:00'
    else:
        return 0
df['TIME_INTERVAL'] = df['TIME'].apply(conditions)
```

```python
df.head()
```

```python
#Create a turnstile column as identifier
df['TURNSTILE'] = df['STATION'] + "_" + df['C/A'] + "_" + df['UNIT'] + "_" + df['SCP']
```

```python
df.head()
```

```python
mta_df = df.copy()
```

```python
mta_df.head()
```

```python
#Create previous' datetime's cumulative entries
d = mta_df.set_index([mta_df.DATETIME,'TURNSTILE'])['ENTRIES'].to_dict()
d2 = mta_df.set_index([mta_df.DATETIME,'TURNSTILE'])['EXITS'].to_dict()
mta_df['PREVIOUS_ENTRIES'] = (mta_df.set_index([mta_df.DATETIME-timedelta(hours=4),'TURNSTILE']).index.to_series().map(d).reset_index(drop=True))
mta_df['PREVIOUS_EXITS'] = (mta_df.set_index([mta_df.DATETIME-timedelta(hours=4),'TURNSTILE']).index.to_series().map(d2).reset_index(drop=True))
```

```python
mta_df.head()
```

```python
mta_df.info()
```

```python
#Drop NaN's row
mta_df = mta_df.dropna()
```

```python
#Analyze counter
mta_df[mta_df['ENTRIES']<mta_df['PREVIOUS_ENTRIES']].head()
```

```python
#Analyze counter
mta_df[mta_df['EXITS']<mta_df['PREVIOUS_EXITS']].head()
```

```python
#Fix counter
def get_periodic_entries(df):
    counter = df['PREVIOUS_ENTRIES'] - df['ENTRIES']
    if counter < 0:
        counter = -counter
    else:
        pass
    return counter

def get_periodic_exits(df):
    counter = df['PREVIOUS_EXITS'] - df['EXITS']
    if counter < 0:
        counter = -counter
    else:
        pass
    return counter

#Add periodic entries and exits
mta_df['PERIODIC_ENTRIES'] = mta_df.apply(get_periodic_entries,axis=1)
mta_df['PERIODIC_EXITS'] = mta_df.apply(get_periodic_exits,axis=1)
```

```python
mta_df.head()
```

```python
#Drop unnecessary columns
mta_df = mta_df.drop(["PREVIOUS_ENTRIES","PREVIOUS_EXITS"],axis=1)
```

```python
#Create a column illustrates the total number of people, who visited the station
mta_df['TOTAL_PEOPLE'] = mta_df['PERIODIC_ENTRIES'] + mta_df['PERIODIC_EXITS']
```

```python
mta_df.head()
```

```python
#Drop unnecessary columns
mta_df = mta_df.drop(['PERIODIC_ENTRIES','PERIODIC_EXITS'],axis=1)
```

```python
mta_df.head()
```

```python
turnstile_based_analysis = mta_df.groupby(['C/A','UNIT','SCP','STATION'])[['TOTAL_PEOPLE']].sum().copy()
turnstile_based_analysis.sort_values('TOTAL_PEOPLE',ascending=False).head(10)
```

```python
day_based_analysis = mta_df.groupby(['DAY'])[['TOTAL_PEOPLE']].sum().copy()
day_based_analysis.sort_values('TOTAL_PEOPLE',ascending=False).head(10)
```

```python
date_based_analysis = mta_df.groupby(['DATE'])[['TOTAL_PEOPLE']].sum().copy()
date_based_analysis.sort_values('TOTAL_PEOPLE',ascending=False).head(10)
```

```python
time_based_analysis = mta_df.groupby(['TIME'])[['TOTAL_PEOPLE']].sum().copy()
time_based_analysis.sort_values('TOTAL_PEOPLE',ascending=False).head(10)
```

```python
datetime_based_analysis = mta_df.groupby(['DATETIME'])[['TOTAL_PEOPLE']].sum().copy()
datetime_based_analysis.sort_values('TOTAL_PEOPLE',ascending=False).head(10)
```
