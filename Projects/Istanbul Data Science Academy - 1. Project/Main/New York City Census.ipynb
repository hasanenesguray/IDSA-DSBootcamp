```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import datetime
from functools import partial, reduce

#Data import and configuration
df = pd.read_csv('/Users/hasanenesguray/Desktop/Bootcamp/Projects/1. Project/ACSST1Y2021.S0101-2022-10-01T102940.csv')
df2 = df[df['Label (Grouping)']=='Total population']
female_pop = df2[['Bronx borough, Bronx County, New York!!Female!!Estimate',
                  'Staten Island borough, Richmond County, New York!!Female!!Estimate',
                 'Brooklyn borough, Kings County, New York!!Female!!Estimate',
                 'Manhattan borough, New York County, New York!!Female!!Estimate',
                 'Queens borough, Queens County, New York!!Female!!Estimate']]

pd.options.mode.chained_assignment = None  # default='warn'
female_pop.rename(columns={'Bronx borough, Bronx County, New York!!Female!!Estimate': 'Bronx'},inplace=True)
female_pop.rename(columns={'Staten Island borough, Richmond County, New York!!Female!!Estimate': 'Staten Island'},inplace=True)
female_pop.rename(columns={'Brooklyn borough, Kings County, New York!!Female!!Estimate': 'Brooklyn'},inplace=True)
female_pop.rename(columns={'Manhattan borough, New York County, New York!!Female!!Estimate': 'Manhattan'},inplace=True)
female_pop.rename(columns={'Queens borough, Queens County, New York!!Female!!Estimate': 'Queens'},inplace=True)

female_pop = female_pop.transpose()
female_pop.rename(columns={0: 'Total Female Population'},inplace=True)

def borough_identifier(x):
    return x
def integer_converter(x):
    x =str(x)
    x = x.replace(",","")
    try:
        return int(x)
    except:
        return None
    
female_pop['Borough'] = female_pop.index.to_series().apply(borough_identifier)
female_pop['Total Female Population'] = female_pop['Total Female Population'].apply(integer_converter)

#Plot graph for Female Population
plt.figure(figsize=(15,8))
plt.title('Female Population in New Your Boroughs (2022 Estimation)',fontsize = 30,loc='CENTER',color='black',weight='bold');
plt.xlabel('x',weight='bold',fontsize=20)
plt.ylabel('y',weight='bold',fontsize=20)
sns.barplot(data = female_pop, y = female_pop['Total Female Population'], x = female_pop.Borough)
plt.show()
```
